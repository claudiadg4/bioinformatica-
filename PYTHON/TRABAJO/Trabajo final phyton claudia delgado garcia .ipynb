#### 5.- Devuelve la secuencia de ARN que surgiría a partir de la siguiente secuencia. Recuerda que la secuencia de ARN resultante tiene que ser complementaria a la de ADN. (1 pto)

```python
#Para crear una funcion que corresponda con el arn complememtario a mi seciencia de adn
def dna_to_rna_complement(dna_seq):
    # Creo un diccionario de complementariedad ADN -> ARN
    complement = {
        "A": "U",  # A en ADN = U en ARN
        "T": "A",  # T en ADN = A en ARN
        "G": "C",  # G en ADN = C en ARN
        "C": "G"   # C en ADN = G en ARN
    }
    # Para generar la secuencia complementaria de ARN
    rna_seq = "".join(complement[base] for base in dna_seq)
    return rna_seq

# Con esto conseguimos aplicar la función a mi secuencia del gen CNTNAP2
rna = dna_to_rna_complement(secuencia)

# Para mostrar los resultados
print("\nPrimeros 200 nucleótidos de la secuencia de ARN:")
print(rna[:200])
print("\nLongitud de la secuencia de ARN:", len(rna), "nucleótidos")
```

```python
# Esta es una función para la digestión de ADN con enzimas de restriccion
def digestion_restriccion(secuencia, enzimas):
    # Limpiamos la secuencia, es decir evitamos saltos de línea, espacios y ponemos la secuencia en mayusculas.
    seq = secuencia.replace("\n", "").replace(" ", "").upper()
    
# Para las enzimas que cortan
    cortes = {} 
# Enzs las que no cortan
    no_cortan = []     
# Fragmentos  
    fragmentos = []   

#para buscar los sitios de corte de cada enzima
#utilizamos el * para determinar el sitio de corte
    for nombre, patron in enzimas.items():
        corte_idx = patron.index('*')
        sitio = patron.replace('*', '')  # sitio  reconocimiento sin asterisco
        
        posiciones = []
        start = 0
        while True:
            pos = seq.find(sitio, start)
            if pos == -1:
                break
            posiciones.append(pos + corte_idx)  # posición real de los cortes
            start = pos + 1
        if posiciones:
            cortes[nombre] = posiciones
        else:
            no_cortan.append(nombre)

    # para similar la digestion con fragmentos ADN 
    posiciones_corte = sorted([p for lista in cortes.values() for p in lista])
    inicio = 0
    for pos in posiciones_corte:
        fragmentos.append(seq[inicio:pos])
        inicio = pos
    fragmentos.append(seq[inicio:])  # Último fragmento

     # para mostrar los resultados
    print("Enzimas que cortan y posiciones:")
    for enzima, pos in cortes.items():
        print(f" - {enzima}: {pos}")
    
    if no_cortan:
        print("\nEnzimas que no cortan:", ", ".join(no_cortan))
    else:
        print("\nTodas las enzimas tienen al menos un sitio de corte.")

    print(f"\nNúmero total de fragmentos: {len(fragmentos)}")
    print("Tamaños de los primeros 10 fragmentos:", [len(f) for f in fragmentos[:10]])

    # Para devolver
    return cortes, no_cortan, fragmentos
    # Para llamar a la función secuencia de ADN
cortes, no_cortan, fragmentos = digestion_restriccion(secuencia, enzR)
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
# Creo un diccionario de codones para poder traducir de ADN a proteína
codon_table = {
    'ATA':'I','ATC':'I','ATT':'I','ATG':'M',
    'ACA':'T','ACC':'T','ACG':'T','ACT':'T',
    'AAC':'N','AAT':'N','AAA':'K','AAG':'K',
    'AGC':'S','AGT':'S','AGA':'R','AGG':'R',
    'CTA':'L','CTC':'L','CTG':'L','CTT':'L',
    'CCA':'P','CCC':'P','CCG':'P','CCT':'P',
    'CAC':'H','CAT':'H','CAA':'Q','CAG':'Q',
    'CGA':'R','CGC':'R','CGG':'R','CGT':'R',
    'GTA':'V','GTC':'V','GTG':'V','GTT':'V',
    'GCA':'A','GCC':'A','GCG':'A','GCT':'A',
    'GAC':'D','GAT':'D','GAA':'E','GAG':'E',
    'GGA':'G','GGC':'G','GGG':'G','GGT':'G',
    'TCA':'S','TCC':'S','TCG':'S','TCT':'S',
    'TTC':'F','TTT':'F','TTA':'L','TTG':'L',
    'TAC':'Y','TAT':'Y','TAA':'_','TAG':'_',
    'TGC':'C','TGT':'C','TGA':'_','TGG':'W'
}

def analizar_ORFs(secuencia):
    # Hebra complementaria inversa
    def complementaria(seq):
        comp = {'A':'T','T':'A','C':'G','G':'C'}
        return ''.join(comp[b] for b in seq)

    # esto es la traducción de ADN a proteína
    def traducir(seq):
        prot = ""
        for i in range(0, len(seq) - 2, 3):
            codon = seq[i:i+3]
            prot += codon_table.get(codon, "X")
        return prot

    # Para preparar las 6 posibles lecturas
    strands = {
        '+': secuencia,
        '-': complementaria(secuencia)[::-1]
    }

    orfs = []

    # para buscar ORFs en cada marco de lectura
    for strand_name, seq in strands.items():
        for frame in range(3):
            prot = traducir(seq[frame:])
            i = 0
            while i < len(prot):
                if prot[i] == "M":
                    j = i
                    while j < len(prot) and prot[j] != "_":
                        j += 1
                    orfs.append({
                        'protein': prot[i:j],
                        'strand': strand_name
                    })
                    i = j
                else:
                    i += 1

    # aqui ordeno por longitud descendente
    orfs.sort(key=lambda x: len(x['protein']), reverse=True)

    # para quedarme con los 6 mas largos
    top6 = orfs[:6]

    # la mejor predicción, es decir el ORF mas largo
    mejor = top6[0]

    # Cálculo el porcentaje de los aminoacidos aromaticos que hay
    arom = sum(mejor['protein'].count(a) for a in ['F','Y','W'])
    mejor['porcentaje_aromaticos'] = (arom / len(mejor['protein'])) * 100

    # Calculo la longitud relativa de los aminoacidos aromaticos comparada con el total del ADN
    mejor['longitud_relativa'] = (len(mejor['protein']) * 3) / len(secuencia) * 100

    return orfs, top6, mejor

    
# Llamo a la función de la secuencia 
orfs, top6, mejor_proteina = analizar_ORFs(secuencia)

# para mostrar los 6 ORFs más largos
print("6 ORFs más largos:")
for i, o in enumerate(top6, 1):
    print(f"{i}. Longitud: {len(o['protein'])} aa, Strand: {o['strand']}")

# para mostrar la mejor predicción
if mejor_proteina:
    print("\nMejor predicción (ORF más largo):")
    print("Longitud:", len(mejor_proteina['protein']), "aa")
    print("Strand:", mejor_proteina['strand'])
    print("Porcentaje de aminoácidos aromáticos:", round(mejor_proteina['porcentaje_aromaticos'],2), "%")
    print("Longitud relativa respecto al ADN:", round(mejor_proteina['longitud_relativa'],2), "%")
```
